```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

# Define Cooling Model

```python
def cooling_model(t, T_ambient, T_initial, k):
    return T_ambient + (T_initial - T_ambient) * np.exp(-k * t)
```

# Apply Cooling Model to data

```python
time_extended = np.linspace(0, 30, 500)

temperature_projectedwet = cooling_model(time_extended, T_ambient = 70, T_initial = 921, k=0.28)
temperature_projecteddry = cooling_model(time_extended, T_ambient = 140, T_initial = 950, k=0.335)
temperature_projectedpath = cooling_model(time_extended, T_ambient = 110, T_initial = 1160, k=0.3)
temperature_projectedroad = cooling_model(time_extended, T_ambient = 80, T_initial = 1155, k=0.28)
```

# Create Bins for Heatmap

```python
bin_edges = [0, 0.5, 1, 1.5, 2]


wetvdry10 = wetvdry10[(wetvdry10['Hours'] >= bin_edges[0]) & (wetvdry10['Hours'] <= bin_edges[-1])]


time_intervals10 = pd.cut(wetvdry10['Hours'], 
                        bins=[0, 0.5, 1, 1.5, 2], 
                        labels=['00.5 hrs', '0.51 hrs', '11.5 hrs', '1.52 hrs'],
                        include_lowest=True)
wetvdry10['Time Interval'] = time_intervals10
```

# Create Temperature Matrix from data and turn it into a numpy array

```python
depth_columnswet10 = ['contact_10wet', '20_10wet', '40_10wet', '60_10wet']
temperature_matrixwet10 = wetvdry10.groupby('Time Interval')[depth_columnswet].mean()

depth_columnsdry10 = ['contact_10dry', '20_10dry', '40_10dry', '60_10dry']
temperature_matrixdry10 = wetvdry10.groupby('Time Interval')[depth_columnsdry].mean()

temperature_datawet10 = temperature_matrixwet10.to_numpy()
temperature_datadry10 = temperature_matrixdry10.to_numpy()
```

# Flip data to ensure depth is on y axis and apply a log transformation

```python
temperature_datawet_flipped10 = temperature_datawet10.T
temperature_datadry_flipped10 = temperature_datadry10.T

# Apply log transform to wet and dry data
temperature_datawet_log10 = np.log1p(temperature_datawet_flipped10)  # Use log1p to safely handle zeros
temperature_datadry_log10 = np.log1p(temperature_datadry_flipped10)
```

```python


fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Wet soil heatmap with log scale
heatmap_wet = axes[0].imshow(temperature_datawet_log10, aspect='auto', cmap='coolwarm')
axes[0].set_xticks(range(len(temperature_matrixwet10.index)), labels=temperature_matrixwet10.index, rotation=45,fontsize=13)
axes[0].set_yticks(range(len(depth_columnswet10)), labels=['Contact', '20 cm', '40 cm', '60 cm'], fontsize=13)
axes[0].set_title('Wet Soil', fontsize=16, fontweight="bold")
axes[0].set_xlabel('Time Intervals', fontsize=14)
axes[0].set_ylabel('Depth below Lava Flow',fontsize=14)


# Dry soil heatmap with log scale
heatmap_dry = axes[1].imshow(temperature_datadry_log10, aspect='auto', cmap='coolwarm')
axes[1].set_xticks(range(len(temperature_matrixdry10.index)), labels=temperature_matrixdry10.index, rotation=45, fontsize=13)
axes[1].set_yticks(range(len(depth_columnsdry10)), labels=['Contact', '20 cm', '40 cm', '60 cm'], fontsize =13)
axes[1].set_title('Dry Soil',fontsize=16, fontweight="bold")
axes[1].set_xlabel('Time Intervals', fontsize=14)
axes[1].set_ylabel('Depth below Lava Flow',fontsize=14)

cbar = fig.colorbar(heatmap_wet, ax=axes, orientation='vertical', fraction=0.05, pad=0.2)
cbar.ax.set_position([0.92, 0.15, 0.02, 0.7])  # [x, y, width, height]
cbar.set_label('Soil Temperature (log)(C)', fontsize=14)

plt.suptitle('Temperature Heatmaps for 10cm Lava Flow', fontsize = 24, fontweight='bold')
plt.subplots_adjust(wspace=0.3)

plt.show()
```
